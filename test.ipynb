```python
%matplotlib inline
from matplotlib import pyplot as plt

from My_Mnist_Net.package.net.layer import *
from My_Mnist_Net.package.net.optim import *
```

```python
w = np.array([2])
b = np.array([500])
data_x = np.arange(-100, 100,2)
data_y = data_x ** 2 * w + b + np.random.randn()
plt.scatter(data_x, data_y)
```

```python
net = Net(MSE_loss, SGD(0.05), [Liner(1, 1), SigmoidLayer(), Liner(1, 1)])
for x, y in zip(data_x, data_y):
    res = net.train(x, y)
    print(res.item())
```

```python
pre_y = net.predict(data_x)
plt.scatter(data_x, pre_y)
```
